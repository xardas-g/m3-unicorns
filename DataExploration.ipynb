```python
weatherAUS["WindGustDir"].value_counts()
#%%
directions = ["N", "NNE", "NE", "ENE", "E", "ESE", "SE", "SSE", "S", "SSW", "SW", "WSW", "W", "WNW", "NW", "NNW"]

# Convert directions into degrees
# Each step is 360 / 16 = 22.5 degrees
direction_to_number = {direction: i * 22.5 for i, direction in enumerate(directions)}

numbers = []

# Print the mapping
for direction, number in direction_to_number.items():
    numbers.append(number) 
```

```python
from mlrain.data import Dataset
from mlrain.model_factory import ModelFactory
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
```

```python

# define & load dataset
X, y = Dataset.load_xy()
# X = weatherAUSNumeric.drop(columns=COL_RAINTOMORROWBOOL)
# y = weatherAUSNumeric[COL_RAINTOMORROWBOOL]

# split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.3, shuffle=True)

    # define models to be evaluated
models = [
    ModelFactory.create_logistic_regression_orig(),
    ModelFactory.create_knn_orig(),
    ModelFactory.create_random_forest_orig(),
    ModelFactory.create_decision_tree_orig(),
]

    # evaluate models
for model in models:
    print(f"Evaluating model:\n{model}")
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print(confusion_matrix(y_test, y_pred))
    print(classification_report(y_test, y_pred))
```
